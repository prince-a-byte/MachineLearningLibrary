```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random dataset for testing my model....
from sklearn.datasets import make_regression
X,y = make_regression(
    n_samples=100,
    n_features=2,
    noise=10,
    random_state=12
)
# plt.scatter(X,y,color='blue')
# plt.show()
```

```python
# Using sklearn Metrics to verify my metrics
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test  = train_test_split(X,y,test_size=0.2,random_state=12)
lr = LinearRegression()
lr.fit(X_train,y_train)

y_pred = lr.predict(X_test)

MAE = mean_absolute_error(y_test,y_pred)
print(f"Mean Absolute Error is:- {MAE}")
```

```python
# Using my own metrics
def mean_absolute_errorr(y_test,y_pred):
    MAE = 0
    for i in range(y_test.shape[0]):
        if((y_test[i] - y_pred[i]) >= 0):
            MAE += ((y_test[i] - y_pred[i])) / y_test.shape[0]

        else:
            MAE += ((y_pred[i] - y_test[i])) / y_test.shape[0]
    return MAE

MAE1 = mean_absolute_errorr(y_test,y_pred)
print(f"Mean Absolute Error is:- {MAE1}")
```
